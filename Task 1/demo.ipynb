```python
from src.mnist_classifier import MnistClassifier
from src.utils import load_mnist

```

```python
# Load the MNIST dataset
x_train, y_train, x_test, y_test = load_mnist()
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
# Fast EDAs
import matplotlib.pyplot as plt
import numpy as np

# Plot the first 25 images from the training set
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.imshow(x_train[i], cmap=plt.cm.binary)
    plt.xlabel(y_train[i])
    plt.xticks([])
    plt.yticks([])
```

```python
# Create the classifier
cnn_classifier = MnistClassifier('cnn')
cnn_classifier.train(x_train, y_train)
cnn_predictions = cnn_classifier.predict(x_test)

nn_classifier = MnistClassifier('nn')
nn_classifier.train(x_train, y_train)
nn_predictions = nn_classifier.predict(x_test)

rf_classifier = MnistClassifier('rf')
rf_classifier.train(x_train, y_train)
rf_predictions = rf_classifier.predict(x_test)
```

```python
# Plot the 25 test images with their predictions from the RF, CNN and NN classifiers, with the true labels
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_test[i].reshape(28, 28), cmap=plt.cm.binary)
    plt.xlabel(f"True: {y_test[i]}, CNN: {cnn_predictions[i]}, NN: {nn_predictions[i]}, RF: {rf_predictions[i]}",
               color='black' if y_test[i] == cnn_predictions[i] == nn_predictions[i] == rf_predictions[i] else 'red',
                fontsize=6)
```
