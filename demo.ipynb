# Task 2: Demo
Let's load a random image and make a prediction. Then, we will get the guess. Finally, we will check if the guess is correct.

```python
import PIL
from matplotlib import pyplot as plt

from classification import get_random_image, predict
from ner import get_entities
from pipeline import process_user_input
```

Let's load a random image and make a prediction.

```python
image_path = get_random_image()
# Plot the image
plt.imshow(PIL.Image.open(image_path))
```

Predict the class of the image.

```python
prediction = predict(image_path)
prediction
```

Get the standard output from the user.

```python
text = "There is a cow in the picture."
```

```python
result = process_user_input(text, prediction)
def print_result(result, prediction, text):
    if result:
        print("Correct! It's a {}!".format(prediction))
    else:
        print("Incorrect! This is {}.".format(prediction))
    print(f"Your guess: {get_entities(text)[0]['entity']}")
```

```python
print_result(result, prediction, text)
```
